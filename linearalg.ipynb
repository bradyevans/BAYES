```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors a and b
a = np.array([1, 2])
b = np.array([3, 1])

# Define coefficients c1 and c2
c1 = 2
c2 = 3

# Calculate the linear weighted combination
c1a = c1 * a
c2a = c2 * b
c = c1a + c2a
print(c)
# c = c1 * a + c2 * b

# Create the plot
fig, ax = plt.subplots()

# Plot vectors as arrows
ax.arrow(0, 0, *c1a, head_width=0.2, head_length=0.1, fc='black', ec='black', label='c1*a')
ax.arrow(0, 0, *c2a, head_width=0.2, head_length=0.1, fc='yellow', ec='yellow', label='c2*b')
ax.arrow(0, 0, *a, head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='a')
ax.arrow(0, 0, *b, head_width=0.1, head_length=0.1, fc='green', ec='green', label='b')
ax.arrow(0, 0, *c, head_width=0.1, head_length=0.1, fc='red', ec='red', label='c1*a + c2*b')

# Set axis limits and labels
ax.set_xlim(0, c1*a[0] + c2*b[0] + 1)
ax.set_ylim(0, c1*a[1] + c2*b[1] + 1)
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Define vectors a and b
a = np.array([1, 2])
b = np.array([3, 1])

# Create the plot
fig, ax = plt.subplots()

# Plot vectors as arrows for a and b
ax.arrow(0, 0, *a, head_width=0.5, head_length=0.5, fc='blue', ec='blue', label='a')
ax.arrow(0, 0, *b, head_width=0.5, head_length=0.5, fc='green', ec='green', label='b')

# Generate and plot linear weighted combinations
for _ in range(2000):
    c1 = random.uniform(1, 10)
    c2 = random.uniform(1, 10)
    c = c1 * a + c2 * b
    ax.arrow(0, 0, *c, head_width=0.5, head_length=0.5, fc='red', ec='red', alpha=0.5)

# Set axis limits and labels
max_val = 10 * (max(a[0], b[0]) + max(a[1], b[1]))
ax.set_xlim(0, max_val)
ax.set_ylim(0, max_val)
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Show the plot
plt.show()
```
